### Basic Concatenation
#### How would you concatenate coffee_shops and more_coffee_shops vertically (i.e., stacking them on top of each other)?

```python
pd.concat([coffee_shops, more_coffee_shops])
```

### Ignoring Index
#### If you wanted to reset the index after concatenation so that it runs continuously without repeating, which parameter would you use?

```python
pd.concat([coffee_shops, more_coffee_shops], ignore_index=True)
```

### Adding MultiIndex
#### How would you concatenate coffee_shops and more_coffee_shops in a way that the resulting DataFrame has a multi-index distinguishing between the two original DataFrames?

```python
pd.concat([coffee_shops, more_coffee_shops], keys=['first', 'second'])
```
