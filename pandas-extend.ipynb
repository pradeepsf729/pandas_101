```python
import pandas as pd
```

```python
data = {
'Date': pd.date_range(start='2023-08-01', periods=10, freq='D'),
'Flavor': ['Vanilla', 'Chocolate', 'Strawberry', 'Mint', 'Caramel', 'Cookies n Cream', 'Rocky Road', 'Mango', 'Pistachio', 'Blueberry'],
'Units_Sold': [100, 120, 90, 105, 110, 85, 95, 115, 125, 105],
'Profit': [200, 240, 180, 210, 220, 170, 190, 230, 250, 210]
}

bruams_df = pd.DataFrame(data)
```

```python
bruams_df
```

### Expanding Mean:
#### How would you calculate the expanding mean of Units_Sold across the days?

```python
bruams_df.loc[:, ["Units_Sold"]].expanding().mean()
```

### Profit's Expanding Sum
#### If you want to see the total profit till each day, how would you use the expanding function?

```python
df_copy = bruams_df.copy()
```

```python
df_copy['Profit'] = bruams_df.loc[:, ['Profit']].expanding().sum()
```

```python
df_copy
```

### Max Units Sold:
#### How can you find the maximum Units_Sold until each day using the expanding function?

```python
df_copy['Units_Sold'] = bruams_df.loc[:, ['Units_Sold']].expanding().max()
```

```python
df_copy
```

### Minimum Profit Over Time:
#### Using the expanding function, how would you compute the minimum profit earned until each specific day?

```python
df_copy['Units_Sold'] = bruams_df.loc[:, ['Units_Sold']].expanding().min()
```

```python
df_copy
```

### Calculate Expanding Standard Deviation:
#### If you wanted to understand the volatility in Units_Sold, how would you calculate the expanding standard deviation for it?

```python
df_copy['Units_Sold'] = bruams_df.loc[:, ['Units_Sold']].expanding().std()
```

```python
df_copy
```

### Count Using Expanding:
#### How can you use the expanding function to count the number of days until each specific date?

```python
df_copy['num_days'] = bruams_df.loc[:, ['Date']].expanding().count().astype('int32')
```

```python
df_copy
```

### Expanding Median:
#### How would you calculate the median of Profit using an expanding window up to each day?

```python
df_copy['Profit'] = bruams_df.loc[:, ['Profit']].expanding().median()
```

```python
df_copy
```

### Multiple Expanding Operations:
#### If you want to compute both the expanding mean and expanding standard deviation of Units_Sold in one call, how would you do it?

```python
df_copy[['Units_Sold_mean', 'Units_Sold_std']] = bruams_df.loc[:, ['Units_Sold']].expanding().agg(['mean', 'std'])
```

### Conditionally Using Expanding:
#### If you wanted to find the expanding sum of Units_Sold but only for the days when more than 100 units were sold, how would you apply such a condition?

```python
bruams_df[bruams_df['Units_Sold']>100].loc[:, ['Units_Sold']].expanding().sum()
```
