#### What advantage does using between provide over using relational operators (&, |) when filtering rows in df with 'Age' values between 25 and 30?

```python
# better easy to use as we can use method instead of relation operators
```

```python
data = {'Age': [14, 20, 35, 45, 23, 29, 30, 25]}
df = pd.DataFrame(data)
df
```

```python
df[(df['Age']>= 25) & (df['Age']<=30)]
```

```python
df[df['Age'].between(25, 30)]
```
