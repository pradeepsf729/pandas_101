```python
import pandas as pd
```

#### Data

```python
d = {}
d['A'] = list(range(10)) 
d['B'] = list(range(10, 20))
d['C'] = list(range(20, 30))
d['D'] = list(range(30, 40))

data = pd.DataFrame(d)
```

```python
data
```

### using .loc with DataFrame

#### 1) Basic Indexing: How would you use loc to get all the values in the row where the index is 2?

```python
data.loc[2,:]
```

#### 2) Multiple Columns: How would you use loc to retrieve the values in columns 'A' and 'C' for the rows with indices 1 and 3? 


```python
data.loc[[1,3], ['A', 'C'] ]
```

#### 3) Conditional Filtering: How can you use loc to filter rows where the values in column 'B' are greater than or equal to 13 and only show column 'A' for those rows? 


```python
data.loc[data['B'] > 13, ['A']]
```

#### 4) Row and Column Together: How would you use loc to get the value located in the row with index 4 and column 'B'? 


```python
data
```

```python
data.loc[4, ['B']]
```

#### 5) Modify Values: How would you use loc to change the value in the row where the index is 0 and the column is 'C' to 99?


```python
data.loc[0, ['C']] = 99
```

```python
data
```

#### 6) Range of Rows: How would you use loc to select all rows from index 2 to 5, inclusive, and only columns 'A' and 'B'?

```python
data.loc[2:5, ['A', 'B']]
```

#### 7) Boolean Conditions with Multiple Columns: How can you use loc to filter rows where column 'A' is greater than 4 and column 'B' is less than 20, and then only retrieve these rows for columns 'A' and 'C'?

```python
data.loc[(data['A'] > 4) & (data['B'] < 20), ['A', 'C']]
```

#### 8) Modify Multiple Values: How would you use loc to change all values in column 'C' to 0 where column 'B' is less than 15?

```python
data.loc[data['B']<15, 'D'] = 0
```

```python
data
```

#### 9) Slice Columns: How can you use loc to select all rows and columns from 'B' to 'D'?

```python
data.loc[:, 'B':'D']
```

#### 10) Modify Row Values: How would you use loc to update all the column values in a row with index 3 to a new list of values [10, 20, 30, 40]?

```python
data.loc[3] = [10,20,30,40]
```

```python
data
```

#### 11) Negate Conditions: How would you use loc to select all rows where column 'A' is not equal to 5 and then only retrieve the values for columns 'C' and 'D'?

```python
data.loc[data['A'] != 5, ['C', 'D']]
```

#### 12) Multiple Conditions for Rows and Columns: How would you use loc to select rows where column 'A' is greater than 5 and column 'B' is not equal to 7, and then retrieve columns 'A', 'C', and 'E'?

```python
data.loc[(data['A'] > 5) & (data['B'] != 7), ['A', 'C', 'D']]
```

#### 13) Set Values Based on Another Column: How would you use loc to set the values in column 'D' to be the same as the values in column 'B' for rows where column 'A' is greater than 5?

```python
data.loc[data['A'] > 5, 'D'] = data.loc[data['A'] > 5, 'B']
```

```python
data
```

#### 14) Partial String Matching: If column 'E' contains string values, how would you use loc to filter rows where the values in column 'E' start with the letter 'S', and retrieve only column 'A' for those rows?

```python
data['E'] = list(['Stadium', 'Tackle', 'Score', 'Substitute', 'Match', 'Goalkeeper', 'Soccer', 'Penalty', 'Corner', 'Striker'])
```

```python
data
```

```python
data.loc[data['E'].str.startswith('S'), ['A', 'E']]
```

#### 15) Using Functions in Conditions: How would you use loc to filter rows based on a custom function applied to one of the columns, and then retrieve specific columns for those rows?

```python
def f(x) :
    return x > 5
```

```python
data.loc[data['A'].apply(f), ['A', 'E']]
```
